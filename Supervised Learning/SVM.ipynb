```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_decision_regions
```

```python
df= pd.read_csv("placement.csv")
df.head()
```

```python
sns.scatterplot(x="cgpa",y="resume_score",data=df,hue="placed")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["placed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100
```

```python
sv.score(x_test,y_test)*100
```

```python
!python --version
```

```python

plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

# SVR

```python
df1= pd.read_csv("Placementcgpanpackage.csv")
df1.head()
```

```python
x1 = df1[["cgpa"]]           # Double brackets = keeps it as a DataFrame
y1 = df1["package "]
```

```python
sns.scatterplot(x="cgpa",y="package ",data=df1)# TRAILING SPACE IN PACAKGE
plt.show()
```

```python
from sklearn.svm import SVR
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
sv1=SVR(kernel="poly")
sv1.fit(x1_train,y1_train)
```

```python
sv1.score(x1_train,y1_train)*100
```

```python
sv1.score(x1_test,y1_test)*100
```

```python
sns.scatterplot(x="cgpa",y="package ",data=df1)# TRAILING SPACE IN PACAKGE
plt.plot(df1["cgpa"],sv1.predict(x1),color="red")
plt.show()
```
