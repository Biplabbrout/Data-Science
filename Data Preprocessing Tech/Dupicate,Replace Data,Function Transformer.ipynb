# Replace and Change Data type

```python
df.head()
```

```python
df.info() # dependents contains int64 , but it is object datatype, so replace it
```

```python
df.isnull().sum()
```

```python
df["Dependents"].value_counts() # value_counts()and Value_count are differenet
```

```python
# so convert "3+" to"3", befure that fill null values with mode(as first is object type)
```

```python
df["Dependents"]=df["Dependents"].fillna(df["Dependents"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["Dependents"] = df["Dependents"].replace("3+", "3"),#df["Dependents"].replace("3+", "3",inplace=True) it is showing warning
```

```python
df["Dependents"].value_counts()
```

```python
df["Dependents"]=df["Dependents"].astype("int64")
```

# Function Transformer 

```python
# convert non normal distribution data to distribution data
```

```python
# choose the data which dont have missing values
```

```python
sns.histplot(data=df,x="ApplicantIncome")
plt.show()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
q1= df["CoapplicantIncome"].quantile(0.25)
q3= df["CoapplicantIncome"].quantile(0.75)
```

```python
IQR=q3-q1
IQR
```

```python
min_range= q1-(1.5*IQR) # so the outliers are beyond this data
max_range =q3+(1.5*IQR)
min_range,max_range
```

```python
df=df[df["ApplicantIncome"]<=max_range]
```

```python
df
```

```python
df.shape
```

```python
# after removing outliers
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
ft.fit(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_tf"]=ft.transform(df[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(data=df,x="CoapplicantIncome")
plt.title("Before")

plt.subplot(1,2,2)
sns.histplot(data=df,x="CoapplicantIncome_tf")
plt.title("After")
plt.show()
```

```python
ft1=FunctionTransformer(func=lambda x: pow(x,2))
ft1.fit(df[["CoapplicantIncome"]])
df["CoapplicantIncome_tf1"]=ft1.transform(df[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(data=df,x="CoapplicantIncome")
plt.title("Before")

plt.subplot(1,2,2)
sns.histplot(data=df,x="CoapplicantIncome_tf1")
plt.title("After")
plt.show()
```
