## OrdinalEncoder

```python
# here sklearn is used for encoding
```

```python
df2 = pd.DataFrame({"size":["s","m","s","l","l","xl","s","xl","m","s","l"]}) # here df2 is used so use df2 for analysis
```

```python
df2
```

```python
ord_data =[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df2[["size"]])
```

```python
df2["size_en"]=oe.transform(df2[["size"]]) # we can also use df2["size"]=oe.transform(df2[["size"]]) , it will just replace values with the size ,
#size_en will not be there
```

```python
df2
```

```python
# now we will use map() function for ordinal encoding, the only difference between two is in map() we can give the order according to our choice
```

```python
ord_data1={"s":5,"m":6,"l":7,"xl":8}
```

```python
df2["map_en"] = df2["size"].map(ord_data1)
```

```python
df2
```

```python
en_ord_data=[['Rural','Semiurban','Urban']]
```
